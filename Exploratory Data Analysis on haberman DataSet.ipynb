# cdf

```python
#axil nodes
#for people living more than or equals 5 years
plt.close()
counts,bin_edges=np.histogram(serv1["axil_nodes"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="pdf for people living more than or equals 5 years")
plt.plot(bin_edges[1:],cdf,label="cdf for people living more than or equals 5 years")
plt.legend()
```

```python
#axil nodes
#for people living less than 5 years
counts,bin_edges=np.histogram(serv2["axil_nodes"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label="pdf for people living less than 5 years")
plt.plot(bin_edges[1:],cdf,label="cdf for people living less than 5 years")

plt.legend()
plt.show()
```
